# Brain Network Models

### Import Relevant libraries and files

```python
%run utils.py
import importlib
import model
importlib.reload(model)

from model import HeterodimerModel, FisherKolmogorovModel, X
from utils import plotWrapper

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

X
```

### Initialize the parameters for each model

```python
# Define the parameters for each model
params_HD = {
    'k0': 1.0,
    'k1': 0.5,
    'k12': 0.5,
    'k1_tilda': 0.5,
    'y0_1': [2.0, 0.1],
    'y0_2': [2.0,  0.1, 2.0, 0.0], 
    'D': 1e-4
    
}

params_FK = {
    'alpha': 0.5,
    'y0_1': [0.1],
    'y0_2': [0.1, 0.001], 
    'D': 1e-4
}


start_time = 0
step_size = 0.4
num_steps  = 100
end_time =  start_time + (step_size * num_steps)

t_span = (start_time, end_time)
t_eval = np.linspace(start_time, end_time - step_size, num_steps)

# solution outputs are
# .t - time
# y[0] - u_1
# y[1] - v_1
```

---------------------

# Fisher - KPP Model

### Single Node

```python
fisher_model_1 = FisherKolmogorovModel(**params_FK, nodes = 1)

solution_fisher_model_1 = solve_ivp(fun = fisher_model_1.model, 
                     t_span = t_span, 
                     y0 = fisher_model_1.y0,
                     t_eval = t_eval,
                     method = 'BDF')

sol_fk_1 = solution_fisher_model_1.y[0]

plot_fisher_model_1 = plotWrapper(solution_fisher_model_1.t, [solution_fisher_model_1.y[0]], labels = ['Node 1'], 
                                 xlabel = "Time", ylabel = "Concentration", title = fisher_model_1.modelName)
```

### Two Nodes Connected

```python
fisher_model_2 = FisherKolmogorovModel(**params_FK, nodes = 2)

solution_fisher_model_2 = solve_ivp(fun = fisher_model_2.model, 
                     t_span = t_span, 
                     y0 = fisher_model_2.y0,
                     t_eval = t_eval,
                     method = 'BDF')

plot_fisher_model_2 = plotWrapper(solution_fisher_model_2.t, [solution_fisher_model_2.y[0], solution_fisher_model_2.y[1]], labels = ['Node 1', 'Node 2'], 
                                 xlabel = "Time", ylabel = "Concentration", title = fisher_model_2.modelName)

```

 ------------------



# Heterodimer Model

### Single Node 

```python
heterodimer_model_1 = HeterodimerModel(**params_HD, nodes = 1)

solution_heterodimer_model_1 = solve_ivp(fun = heterodimer_model_1.model, 
                                         t_span = t_span, 
                                         y0 = heterodimer_model_1.y0,
                                         t_eval = t_eval, 
                                         method = 'BDF')

plot_heterodimer_model_1 = plotWrapper(solution_heterodimer_model_1.t, [solution_heterodimer_model_1.y[0], solution_heterodimer_model_1.y[1]], labels = ['Healthy Conc. Node 1', 'Misfolded Conc. Node 1'], 
                                 xlabel = "Time", ylabel = "Concentration", title = heterodimer_model_1.modelName)
```

### Two Nodes Connected

```python
heterodimer_model_2 = HeterodimerModel(**params_HD, nodes = 2)

solution_heterodimer_model_2 = solve_ivp(fun = heterodimer_model_2.model, 
                                         t_span = t_span, 
                                         y0 = heterodimer_model_2.y0,
                                         t_eval = t_eval, 
                                         method = 'BDF')

plot_heterodimer_model_2 = plotWrapper(solution_heterodimer_model_2.t, 
                                  [
                                      solution_heterodimer_model_2.y[0], solution_heterodimer_model_2.y[1], 
                                      solution_heterodimer_model_2.y[2], solution_heterodimer_model_2.y[3]],
                                  labels = ['Healthy Conc. Node 1', 'Misfolded Conc. Node 1', 'Healthy Conc. Node 2', 'Misfolded Conc. Node 2'], 
                                 xlabel = "Time", ylabel = "Concentration", title = heterodimer_model_2.modelName)
```

```python
# comparison of all models in 1 and 2 node cases

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
# Fisher Kolmogorov 1 Node Case
plotWrapper(solution_fisher_model_1.t, [solution_fisher_model_1.y[0]], labels = ['Node 1'], 
                                 xlabel = "Time", ylabel = "Concentration", title = fisher_model_1.modelName)

plt.subplot(2, 2, 2)
# Fisher Kolmogorov 2 Node Case
plotWrapper(solution_fisher_model_2.t, [solution_fisher_model_2.y[0], solution_fisher_model_2.y[1]], labels = ['Node 1', 'Node 2'], 
                                 xlabel = "Time", ylabel = "Concentration", title = fisher_model_2.modelName)

plt.subplot(2, 2, 3)
# Heterodimer 1 Node Case
plot_heterodimer_model_1 = plotWrapper(solution_heterodimer_model_1.t, [solution_heterodimer_model_1.y[0], solution_heterodimer_model_1.y[1]], labels = ['Healthy Conc. Node 1', 'Misfolded Conc. Node 1'], 
                                 xlabel = "Time", ylabel = "Concentration", title = heterodimer_model_1.modelName)

plt.subplot(2, 2, 4)
# Heterodimer 2 Node case
plotWrapper(solution_heterodimer_model_2.t, 
                                  [solution_heterodimer_model_2.y[0], solution_heterodimer_model_2.y[1], 
                                    solution_heterodimer_model_2.y[2], solution_heterodimer_model_2.y[3]],
                                  labels = ['Healthy Conc. Node 1', 'Misfolded Conc. Node 1', 'Healthy Conc. Node 2', 'Misfolded Conc. Node 2'], 
                                 xlabel = "Time", ylabel = "Concentration", title = heterodimer_model_2.modelName)
plt.tight_layout()
plt.show()
```
