```python
import random

def add_random_binary_node_property(G, attr_name="test_js"):
    """
    Adds a new random binary (0 or 1) attribute to all nodes in the graph.

    Parameters
    ----------
    G : networkx.Graph
        The NetworkX graph to modify.
    attr_name : str
        The name of the new node attribute to add.
    """
    for node in G.nodes():
        G.nodes[node][attr_name] = random.choice([10, 1])
    print(f"Added node attribute '{attr_name}' to {G.number_of_nodes()} nodes.")
    return G
```

```python
import networkx as nx

def export_graphml_with_namespace(G, output_path, xmlns_path=None):
    """
    Exports a NetworkX graph to GraphML with proper Gephi-compatible headers.

    Parameters
    ----------
    G : networkx.Graph
        The graph to export.
    output_path : str
        Path to save the GraphML file.
    xmlns_path : str, optional
        Path to your XSD schema file (for example:
        'file:///C:/Users/sranasin/PycharmProjects/NetworkModels/resources/graphml.xsd')
    """
    nx.write_graphml(G, output_path)

    # Patch the <graphml> header and include schema info if provided
    with open(output_path, "r", encoding="utf-8") as f:
        content = f.read()

    if xmlns_path:
        xmlns_header = (
            '<graphml xmlns="http://graphml.graphdrawing.org/xmlns"\n'
            '         xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"\n'
            f'         xsi:schemaLocation="http://graphml.graphdrawing.org/xmlns {xmlns_path}">'
        )
        content = content.replace("<graphml>", xmlns_header)

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(content)

    print(f"GraphML exported to: {output_path}")
```

```python

# 1️⃣ Add simulated node property
G = add_random_binary_node_property(G, "test_js")

# 2️⃣ Export to new GraphML (you can do this later)
export_graphml_with_namespace(
    G,
    output_path=os.path.realpath(os.path.join(base_PATH, "resources/updated_graph.graphml")),
    xmlns_path="file:///C:/Users/sranasin/PycharmProjects/NetworkModels/resources/graphml.xsd"
)
```
